```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('pizzeria.jpg')
dir ='/content/'
os.chdir(dir)
```

```python
lena = cv2.imread('/content/Noisy-Lena-image.png')
plt.imshow(lena)
```

```python
blur_lena = cv2.GaussianBlur(lena,(5,5),0)
plt.subplot(1,2,1)
plt.imshow(lena)
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(blur_lena,cv2.COLOR_BGR2RGB))
```

```python
sharpening_kernel=np.array([[-1,-1,-1],
                             [-1, 9 ,-1],
                             [-1,-1,-1]])
sharpening_image = cv2.filter2D(blur_lena, -1, sharpening_kernel)
```

```python
plt.imshow(sharpening_image)
```

```python
cv2.Canny?
```

```python
edges_original = cv2.Canny(lena,150,150)
edges_blur = cv2.Canny(blur_lena,150,150)
edges_sharpened = cv2.Canny(sharpening_image,150,150)
```

```python
plt.subplot(1,3,1)
plt.imshow(edges_original)
plt.subplot(1,3,2)
plt.imshow(edges_blur)
plt.subplot(1,3,3)
plt.imshow(edges_sharpened)
```

```python
veggie_img=cv2.imread('/content/veggie.png',0)
plt.imshow(veggie_img , cmap='gray')
```

```python
cv2.threshold?
```

```python
ret, thresh1 = cv2.threshold(veggie_img,120,255,cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(veggie_img,120,255,cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(veggie_img,120,255,cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(veggie_img,120,255,cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(veggie_img,120,255,cv2.THRESH_TOZERO_INV)
```

```python
plt.subplot(2,3,1)
plt.imshow(cv2.cvtColor(thresh1,cv2.COLOR_BGR2RGB))
plt.subplot(2,3,2)
plt.imshow(cv2.cvtColor(thresh2,cv2.COLOR_BGR2RGB))
plt.subplot(2,3,3)
plt.imshow(cv2.cvtColor(thresh3,cv2.COLOR_BGR2RGB))
plt.subplot(2,3,4)
plt.imshow(cv2.cvtColor(thresh4,cv2.COLOR_BGR2RGB))
plt.subplot(2,3,5)
plt.imshow(cv2.cvtColor(thresh5,cv2.COLOR_BGR2RGB))
```
